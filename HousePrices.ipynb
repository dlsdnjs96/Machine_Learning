```python
cols = ['RoofMatl', 'Exterior1st', 'RoofStyle', 'HouseStyle', 'Condition1', 'Condition2', 'Neighborhood', 'MasVnrType', 'Foundation', 
        'Exterior2nd', 'GarageType', 'MiscFeature', 'SaleType', 'SaleCondition', 'Electrical', 'Heating', 'BldgType', 'LandSlope', 
        'LotConfig', 'LandContour', 'Alley', 'LotShape', 'Utilities']

for i in cols:
  materials = all_data[i].unique()

  #print('\r\n\r\n##### '+i+' #####')

  mList = [[0 for rows in range(len(materials))] for cols in range(2)]
  mean = 0
  cnt = 0

  for j in materials:
    sum = train.loc[train[i] == j, 'SalePrice'].sum()
    count = train.loc[train[i] == j, 'SalePrice'].count()

    if count > 0:
      ave = round(sum/count)
      mean += ave
      cnt += 1

  mean /= cnt
  all_data[i+'(status)'] = mean

  cnt = 0

  for j in materials:
    sum = train.loc[train[i] == j, 'SalePrice'].sum()
    count = train.loc[train[i] == j, 'SalePrice'].count()

    if count > 0:
      #print(j+' : '+(str)(sum)+' '+(str)(count)+'')

      ave = round(sum/count)
      

      mList[0][cnt] = j
      mList[1][cnt] = ave
      cnt += 1
      all_data.loc[all_data[i] == j, i+'(status)'] = 0


```

```python
x_col = ['WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch', 
       'ScreenPorch', 'MSZoning(status)',
       'ExterQual(status)', 'BsmtQual(status)',
       'BsmtCond(status)', 'BsmtExposure(status)', 'BsmtFinType1(status)',
       'HeatingQC(status)', 'KitchenQual(status)',
       'FireplaceQu(status)', 'GarageFinish(status)', 'GarageQual(status)',
       'GarageCond(status)', 'PoolQC(status)', 'Fence(status)',
       'LotShape(status)', 
       'RoofMatl(status)', 'Exterior1st(status)', 'RoofStyle(status)',
       'HouseStyle(status)', 'Condition1(status)', 
       'Neighborhood(status)', 'MasVnrType(status)', 'Foundation(status)',
       'Exterior2nd(status)', 'GarageType(status)', 
       'SaleType(status)', 'SaleCondition(status)', 'Electrical(status)',
       'Heating(status)', 'BldgType(status)', 
       'LotConfig(status)', 'LandContour(status)', 'CentralAir(status)']
y_col = 'SalePrice'
```

```python
x_train = train[x_col]
y_train = train[y_col]

x_test = test[x_col]

x_train.describe()
```

```python
from xgboost import XGBRegressor
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.linear_model import LinearRegression

from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import mean_squared_error
```

```python
xgb = XGBRegressor(random_state=37)
gbr = GradientBoostingRegressor(random_state=37)
rfr = RandomForestRegressor(random_state=37)

rfc = RandomForestClassifier(random_state=37)
dtc = DecisionTreeClassifier(random_state=37)


models = [xgb, gbr, rfr, rfc, dtc]
names = ['XGBRegressor', 'GradientBoostingRegressor', 'RandomForestRegressor', 'RandomForestClassifier', 'DecisionTreeClassifier']


for md, nm in zip(models, names):
  md.fit(x_train, y_train)
  pred = md.predict(x_test)

  leng = len(y_train)
  if leng > len(pred):
    leng = len(pred)

  print(nm+" : "+(str)(mean_squared_error(y_train[:leng], pred[:leng], squared=False))+"\r\n")
```
