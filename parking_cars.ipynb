```python
all_data = pd.concat([train, test], sort=False)

print(all_data.columns)
all_data.dtypes
```

```python
all_data.loc[all_data['도보 10분거리 내 지하철역 수(환승노선 수 반영)'].isnull(), '도보 10분거리 내 지하철역 수(환승노선 수 반영)'] = 0
all_data.loc[all_data['도보 10분거리 내 버스정류장 수'].isnull(), '도보 10분거리 내 버스정류장 수'] = 0

all_data['지하철'] = all_data['도보 10분거리 내 지하철역 수(환승노선 수 반영)']
all_data['버스'] = all_data['도보 10분거리 내 버스정류장 수']

all_data = all_data.drop('도보 10분거리 내 지하철역 수(환승노선 수 반영)', axis=1, inplace=False)
all_data = all_data.drop('도보 10분거리 내 버스정류장 수', axis=1, inplace=False)

all_data.count()
```

```python
all_data['공가비율'] = 100 * all_data['공가수'] / all_data['총세대수']

all_data = all_data.drop('공가수', axis=1, inplace=False)
#all_data[all_data['임대료'].notnull(), '임대료'] = pd.to_numeric(all_data[all_data['임대료'].notnull(), '임대료'])

all_data.describe()
```

```python
all_data.dtypes
```

```python
array = ['버스', '지하철']

for i in array:
  std = all_data[i].std()/2
  min = all_data[i].min()
  idx = 1

  for j in range(0, 12):
    all_data.loc[all_data[i] >= min, i+'(status)'] = idx
    min += std
    idx += 1

  all_data = all_data.drop(i, axis=1, inplace=False)

```

```python
all_data.columns
```

```python
array = ['임대건물구분', '지역', '공급유형', '자격유형']

for i in array:
  tmp = all_data[i].unique()

  for j in tmp:
    all_data[(str)(i)+'_'+(str)(j)+'(1hot)'] = all_data[i] == j

  all_data = all_data.drop(i, axis=1, inplace=False)

all_data
```
